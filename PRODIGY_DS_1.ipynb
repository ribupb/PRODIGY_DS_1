<h3>Problem statement -</h3>
<p>Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages and genders in a population.</p>
<h3>Dataset used - </h3><p>Employee dataset</p>

<h3>Description -</h3><br>
<li>Continuous variable(Age) - Exploring the histogram of the age distribution in an employee dataset reveals the concentration of employees within specific age ranges and allowing us to identify the most common age groups. It provides insights into the overall age diversity and age-related trends in the workforce.<br>
<li>Categorical variable(Gender) - Exploring the gender breakdown using the histogram and getting the counts of male and female accross the population in the Employee dataset.</li>

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#continuous data

plt.figure(figsize=(9, 6))
plt.hist(df['Age'], bins=17, color='lightgreen', edgecolor='black')
plt.grid(True, linestyle='--', alpha=0.5, color='gray')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.xlabel('Age', fontsize=12, fontweight='bold')  
plt.ylabel('Frequency', fontsize=12, fontweight='bold')  
plt.title('Histogram of Age Distribution', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
#Categorical data

gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['skyblue', 'salmon'], edgecolor='black')

plt.xlabel('Gender', fontsize=12, fontweight='bold')
plt.ylabel('Frequency', fontsize=12, fontweight='bold') 
plt.title('Distribution of Gender', fontsize=14, fontweight='bold') 

plt.xticks(rotation=0, fontsize=10)

for i, count in enumerate(gender_counts):
    plt.text(i, count + 0.1, str(count), ha='center', fontsize=10)

plt.grid(axis='y', linestyle='--', alpha=0.5, color='gray')
plt.tight_layout()
plt.show()
```
